> Question: Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years. Then create a horizontal bar chart showing total runs scored in their debut 5 years.

```python
import matplotlib.pyplot as plt
```

```python
years = [2009, 2010, 2011, 2012, 2013]
sachin_runs = [1743, 1562, 579, 317, 0]
kohli_runs = [1029, 995, 1381, 1028, 1268]
sehwag_runs = [1422, 544, 447, 439, 158]
```

```python
bar_width = 0.25
sachin_pos = [0, 1, 2, 3, 4]
kohli_pos = [x + bar_width for x in sachin_pos]
sehwag_pos = [x + bar_width for x in kohli_pos]
```

```python
plt.bar(sachin_pos, sachin_runs, width=bar_width, label='Sachin')
plt.bar(kohli_pos, kohli_runs, width=bar_width, label='Kohli')
plt.bar(sehwag_pos, sehwag_runs, width=bar_width, label='Sehwag')
plt.ylabel('Runs Scored')
plt.title('Comparison of Runs Scored')
plt.xticks([r + bar_width for r in range(len(years))], years)
plt.legend()
plt.show()
```
