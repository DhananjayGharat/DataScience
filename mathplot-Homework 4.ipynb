```python
import matplotlib.pyplot as plt
```

# Data for plotting

```python
students = ['Omkar', 'Dhananjay', 'Prathamesh', 'Yuvraj', 'Hitesh']
marks = [99, 97, 85, 92, 88]
plt.plot(students,marks)
plt.show()
```

# Adding X and Y Labels

```python
plt.plot(students,marks)
plt.xlabel("Student Names")
plt.ylabel("Marks Scored")
plt.title("Result of the students")
plt.show()
```

# Multiple Lines in One Plot

```python
students = ["Test 1", "Test 2", "Test 3", "Test 4"]
Omkar = [99, 98, 97, 99]
Dhananjay = [87, 92, 89, 90]
plt.plot(students, Omkar, label="Omkar")
plt.plot(students, Dhananjay, label="Dhananjay")
plt.xlabel("Tests")
plt.ylabel("Marks Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

# Using Format Strings

```python
plt.plot(students, Omkar, 'ro--',label="Omkar")
plt.plot(students, Dhananjay, 'g^:',label="Dhananjay")
plt.legend()
```

# Color and Line Style Arguments

```python
Hitesh = [95, 90, 93, 92] 
plt.plot(students, Omkar, color='Magenta', linestyle='--', label="Omkar")
plt.plot(students, Dhananjay, color='green', linestyle='-.', label="Dhananjay")
plt.plot(students, Hitesh, color='blue', label="Hitesh")
plt.xlabel("Tests")
plt.ylabel("Marks Scored")
plt.title("Performance Comparison of Students")
plt.legend()
plt.grid(True)
plt.show()
```

# Line Width and Layout Tweaks

```python
plt.plot(students, Omkar, linewidth=2, label="Omkar")
plt.plot(students, Dhananjay, linewidth=5, label="Dhananjay")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Using Styles in Matplotlib

```python
plt.style.use('fast') 
plt.plot(students, Omkar, linewidth=3, label="Omkar")
plt.plot(students, Dhananjay, linewidth=2, label="Dhananjay")
plt.xlabel("Tests")
plt.ylabel("Marks Scored")
plt.title("Performance Comparison")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

# XKCD Comic Style

```python
with plt.xkcd():
    plt.plot(students, Omkar, label="Omkar")
    plt.plot(students, Dhananjay, label="Dhananjay")
    plt.xlabel("Tests")
    plt.ylabel("Marks Scored")
    plt.title("Performance in XKCD Style")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
