## 1) Plot a Line Chart
Plot y = 3x + 2 for x in the range 0 to 10. Set the x-axis and y-axis labels, and give the plot a title.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure()
x = np.arange(0, 11)
y = 3 * x + 2
plt.plot(x, y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot of y = 3x + 2")
plt.show()
```

## 2) Scatter Plot with Customization
Create a scatter plot of 50 random points. Change the color of the points, increase their size, and add a grid to the plot.

## 3) Bar Chart with Annotations
Draw a vertical bar chart for five categories of your choice. Annotate the value on top of each bar.

```python
import matplotlib.pyplot as plt
plt.figure()
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 24, 10, 32, 8]
bars = plt.bar(categories, values)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')
plt.show()
```

## 4) Multiple Lines in One Plot
Plot three mathematical functions (sin, cos, exp) on the same axes, each with different colors and line styles. Add a legend.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure()
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), 'r--', label='sin(x)')
plt.plot(x, np.cos(x), 'g:', label='cos(x)')
plt.plot(x, np.exp(-x/10), 'b-', label='exp(-x/10)')
plt.legend()
plt.show()
```

## 5) Histogram with Custom Bins

## 6) Subplots Grid
Create a 2x2 grid of subplots: line plot, bar chart, scatter plot, and histogram, each in a different subplot. Add an overall (supertitle) for the figure.

```python
import matplotlib.pyplot as plt
import numpy as np
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Overall Figure Title (Super title)')
x_line = np.arange(0, 11)
y_line = 3 * x_line + 2
axs[0, 0].plot(x_line, y_line)
axs[0, 0].set_title('Line Plot')
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 24, 10, 32, 8]
axs[0, 1].bar(categories, values)
axs[0, 1].set_title('Bar Chart')
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
axs[1, 0].scatter(x_scatter, y_scatter)
axs[1, 0].set_title('Scatter Plot')
data_hist = np.random.randn(1000)
axs[1, 1].hist(data_hist, bins=30)
axs[1, 1].set_title('Histogram')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## 7) Stack Plot
Plot the cumulative sales of 3 products over 12 months as a stack plot using made-up data.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure()
months = np.arange(1, 13)
product1_sales = np.array([10, 12, 15, 18, 20, 22, 25, 28, 30, 32, 35, 38])
product2_sales = np.array([5, 8, 10, 12, 15, 18, 20, 22, 25, 28, 30, 32])
product3_sales = np.array([8, 10, 12, 15, 18, 20, 22, 25, 28, 30, 32, 35])
plt.stackplot(months, product1_sales, product2_sales, product3_sales, labels=['Product 1', 'Product 2', 'Product 3'])
plt.legend(loc='upper left')
plt.xlabel("Month")
plt.ylabel("Cumulative Sales")
plt.title("Cumulative Sales of 3 Products Over 12 Months")
plt.show()
```

## 8) Boxplot Using DataFrame
Given a DataFrame of student exam scores for three subjects, visualize the data distribution with a boxplot.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure()
data = {'Math': [80, 85, 90, 75, 95],
        'Science': [70, 75, 80, 65, 85],
        'English': [90, 95, 85, 80, 75]}
df = pd.DataFrame(data)
sns.boxplot(data=df)
plt.title("Student Exam Scores")
plt.show()
```

## 9) Pie Chart with Explode
Create a pie chart showing the market share of four companies. Explode the largest segment and display percentage values on the chart.

```python
import matplotlib.pyplot as plt
plt.figure()
companies = ['Company A', 'Company B', 'Company C', 'Company D']
market_share = [40, 30, 15, 15]
explode = (0.1, 0, 0, 0)
plt.pie(market_share, explode=explode, labels=companies, autopct='%1.1f%%', startangle=90)
plt.title('Market Share of Four Companies')
plt.axis('equal')
plt.show()
```

## 10) Axis and Legend Customization
For a random line plot, reverse the x-axis, move the legend outside the plot, and increase legend font size.

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(0, 10, 100)
y = np.random.rand(100)
ax.plot(x, y, label='Random Data')
ax.invert_xaxis()
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0., fontsize=12)
plt.show()
```
