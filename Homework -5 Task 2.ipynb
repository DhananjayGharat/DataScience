> Question: Create a pie chart showing the market share of mobile OS (Android, iOS, others). Then recreate the same using a horizontal bar chart and observe which is easier to understand.

```python
import matplotlib.pyplot as plt
```

```python
os_names = ['Android', 'iOS', 'Others']
market_shares = [72.13, 27.48, 0.39]
```

```python
plt.pie(market_shares, labels=os_names, autopct='%1.1f%%')
plt.title('Mobile OS Market Share')
plt.show()
```

>Horizontal Bar Chart

```python
plt.barh(os_names, market_shares)
plt.xlabel('Market Share (%)')
plt.title('Mobile OS Market Share')
plt.show()
```

# Observation -

>Bar Chart is more easier to understand Market Shares
